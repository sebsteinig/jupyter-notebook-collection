# rotate present-day eNd field back to 55Ma using the Herold et al. (2014) rotation file

```python
work_dir       = './'
data_dir       = work_dir
```

```python
# load packages
import numpy as np
import pandas as pd
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.path as mpath
import matplotlib.gridspec as gridspec

import cmocean
import math
import sys
np.set_printoptions(threshold=sys.maxsize)

from netCDF4 import Dataset

from cartopy.util import add_cyclic_point
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
from cartopy import crs

mpl.rcParams["figure.dpi"] = 300
```

# load and plot present-day data

```python
# load present-day data
eNd_data = Dataset(data_dir + 'continentalEpsilonNd.nc', 'r')
eNd_pd = eNd_data.variables['continental_end'][:] 
lats_pd = eNd_data.variables['lat'][:]  # extract/copy the data
lons_pd = eNd_data.variables['lon'][:]
```

```python
# plot present-day data

fig1 = plt.figure(figsize=(15, 10))

gs = gridspec.GridSpec(3, 4)
#gs.update(wspace=0.5)

ax1 = plt.subplot(gs[0, :], projection=ccrs.PlateCarree())
ax1.coastlines()
ax1.title.set_text('present-day continental eNd / Robinson et al. (2021)')
cf1 = ax1.contourf(lons_pd, lats_pd, eNd_pd, 60,transform=ccrs.PlateCarree(), cmap='viridis', vmin = -40, vmax = 10)
plt.colorbar(cf1, label="eNd", orientation="vertical")
gl1 = ax1.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linewidth=1, color='gray', alpha=0.5, linestyle='--')
gl1.xlabels_top = False
plt.show()

ax2 = plt.subplot(gs[1:, :2], projection=ccrs.NorthPolarStereo())
ax2.set_extent([-180, 180, 55, 90], crs=ccrs.PlateCarree())
ax2.coastlines()
ax2.title.set_text('')
cf2 = ax2.contourf(lons_pd, lats_pd, eNd_pd, 60,transform=ccrs.PlateCarree(), cmap='viridis', vmin = -40, vmax = 10)
theta = np.linspace(0, 2*np.pi, 100)
center, radius = [0.5, 0.5], 0.5
verts = np.vstack([np.sin(theta), np.cos(theta)]).T
circle = mpath.Path(verts * radius + center)
ax2.set_boundary(circle, transform=ax2.transAxes)
gl2 = ax2.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linewidth=0.5, color='gray', alpha=0.5, linestyle='--')

ax3 = plt.subplot(gs[1:, 2:], projection=ccrs.SouthPolarStereo())
ax3.set_extent([-180, 180, -90, -55], crs=ccrs.PlateCarree())
ax3.coastlines()
ax3.title.set_text('')
cf3 = ax3.contourf(lons_pd, lats_pd, eNd_pd, 60,transform=ccrs.PlateCarree(), cmap='viridis', vmin = -40, vmax = 10)
theta = np.linspace(0, 2*np.pi, 100)
center, radius = [0.5, 0.5], 0.5
verts = np.vstack([np.sin(theta), np.cos(theta)]).T
circle = mpath.Path(verts * radius + center)
ax3.set_boundary(circle, transform=ax3.transAxes)
gl3 = ax3.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linewidth=0.5, color='gray', alpha=0.5, linestyle='--')
```

# apply paleorotation file to (55Ma from Herold et al. 2014)

```python
# load Herold et al. (2014) rotation file
# this is a 55Ma map showing from which lat-lon any point 'came' going back from present day
eocene_rotation_data = Dataset(data_dir + 'LatLon_55Ma_PD_Herold2014.nc', 'r')
latitudes_eocene_rotation = eocene_rotation_data.variables['latitude'][:] 
longitudes_eocene_rotation = eocene_rotation_data.variables['longitude'][:] 
LAT_eocene_origin = eocene_rotation_data.variables['LAT'][:] 
LON_eocene_origin = eocene_rotation_data.variables['LON'][:] 

# coppy dummy data
# the data will have the same coordinates (lats_pd/lons_pd) as for the present-day
eNd_eocene = eNd_pd.copy()

#print(eNd_pd)
# need to loop over each grid point, look up the respective present-day location in rotation file 
# and assign respective eNd value at the present-day location
for jj in range(len(lats_pd)):
    # find present-day latitude
    lat_idx = (np.abs(latitudes_eocene_rotation - lats_pd[jj])).argmin()
    closest_lat = latitudes_eocene_rotation[lat_idx]
    lat_offset = lats_pd[jj] - closest_lat
    for ii in range(len(lons_pd)):
        # find present-day longitude
        lon_idx = (np.abs(longitudes_eocene_rotation - lons_pd[jj])).argmin()
        closest_lon = longitudes_eocene_rotation[lon_idx]
        lon_offset = lons_pd[jj] - closest_lon
        
        # look up where eocene coordinate 'came' from
        lat_origin = LAT_eocene_origin[lat_idx,lon_idx] + lat_offset
        lon_origin = LON_eocene_origin[lat_idx,lon_idx] + lon_offset
        
        if (math.isnan(lat_origin) or math.isnan(lon_origin)):
            
            eNd_eocene[jj, ii] = -9999
        
        else:

            # find closest coordinates in present-day eNd field
            lat_idx_origin = (np.abs(lats_pd - lat_origin)).argmin()
            lon_idx_origin = (np.abs(lons_pd - lon_origin)).argmin()

            # assign respective present-day eNd value to Eocene field
            eNd_eocene[jj, ii] = eNd_pd[lat_idx_origin,lon_idx_origin]
            
            
            
           # print(eNd_eocene[jj, ii])



```

```python
# load Herold et al. (2014) land-sea-mask
eocene_topo_data = Dataset(data_dir + 'herold_etal_eocene_topo_1x1.nc', 'r')
eocene_topo = eocene_topo_data.variables['topo'][:] 
lats_eocene_topo = eocene_topo_data.variables['lat'][:]
lons_eocene_topo = eocene_topo_data.variables['lon'][:]

# create land-sea-mask from topography field
eocene_lsm = eocene_topo
eocene_lsm[eocene_lsm<0] = 0
eocene_lsm[eocene_lsm>0] = 1
```

```python
# plot rotated Eocene data

fig2 = plt.figure(figsize=(15, 10))

gs = gridspec.GridSpec(3, 4)
#gs.update(wspace=0.5)

ax1 = plt.subplot(gs[0, :], projection=ccrs.PlateCarree())
ax1.title.set_text('Eocene (55Ma) continental eNd / Robinson et al. (2021)')
cf1 = ax1.contourf(lons_pd, lats_pd, eNd_eocene, 60,transform=ccrs.PlateCarree(), cmap='viridis', vmin = -40, vmax = 10)
plt.colorbar(cf1, label="eNd", orientation="vertical")
gl1 = ax1.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linewidth=1, color='gray', alpha=0.5, linestyle='--')
gl1.xlabels_top = False

cl1 = ax1.contour(lons_eocene_topo, lats_eocene_topo, eocene_lsm,transform=ccrs.PlateCarree(), colors="black", linewidths=1.0, levels=[0.5])

plt.show()

ax2 = plt.subplot(gs[1:, :2], projection=ccrs.NorthPolarStereo())
ax2.set_extent([-180, 180, 55, 90], crs=ccrs.PlateCarree())
ax2.title.set_text('')
cf2 = ax2.contourf(lons_pd, lats_pd, eNd_eocene, 60,transform=ccrs.PlateCarree(), cmap='viridis', vmin = -40, vmax = 10)
theta = np.linspace(0, 2*np.pi, 100)
center, radius = [0.5, 0.5], 0.5
verts = np.vstack([np.sin(theta), np.cos(theta)]).T
circle = mpath.Path(verts * radius + center)
ax2.set_boundary(circle, transform=ax2.transAxes)
gl2 = ax2.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linewidth=0.5, color='gray', alpha=0.5, linestyle='--')

cl2 = ax2.contour(lons_eocene_topo, lats_eocene_topo, eocene_lsm,transform=ccrs.PlateCarree(), colors="black", linewidths=1.0, levels=[0.5])

ax3 = plt.subplot(gs[1:, 2:], projection=ccrs.SouthPolarStereo())
ax3.set_extent([-180, 180, -90, -55], crs=ccrs.PlateCarree())
ax3.title.set_text('')
cf3 = ax3.contourf(lons_pd, lats_pd, eNd_eocene, 60,transform=ccrs.PlateCarree(), cmap='viridis', vmin = -40, vmax = 10)
theta = np.linspace(0, 2*np.pi, 100)
center, radius = [0.5, 0.5], 0.5
verts = np.vstack([np.sin(theta), np.cos(theta)]).T
circle = mpath.Path(verts * radius + center)
ax3.set_boundary(circle, transform=ax3.transAxes)
gl3 = ax3.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linewidth=0.5, color='gray', alpha=0.5, linestyle='--')

cl3 = ax3.contour(lons_eocene_topo, lats_eocene_topo, eocene_lsm,transform=ccrs.PlateCarree(), colors="black", linewidths=1.0, levels=[0.5])
```
